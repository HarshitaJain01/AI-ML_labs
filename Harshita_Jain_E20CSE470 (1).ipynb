```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
Train=pd.read_csv('train.csv')
print(Train)
```

```python
Train.describe()
```

```python
print(Train.Pclass.unique())
```

```python
values = {
    'male':Train['Sex'].tolist().count('male'),
    'female':Train['Sex'].tolist().count('female'),
    'mean_fair':Train['Fare'].mean()
}
values
```

```python
fig,ax = plt.subplots(figsize=(20,10))
plt.bar(list(values.keys()),list(values.values()),color='maroon')
plt.xlabel("Men and Women Mean fair")
plt.ylabel("Values")
datas = list(values.values())
for i,data in enumerate(datas):
    plt.text(x=i,y=data+1,s=f"{int(data)}",fontdict=dict(fontsize=10))
plt.tight_layout()
    
plt.show()
```
