```python
def med_target(var):
  temp=Data[Data[var].notnull()]
  temp=temp[[var,'OUTCOME']].groupby(['OUTCOME'])[[var]].median().reset_index()
  return temp
  
```

```python
print(med_target('BMI'))
```

```python
Data.loc[(Data['AGE']<=21), 'Age_Category'] = 'Young'
Data.loc[(Data['AGE']>21) & (Data['AGE']<50), 'Age_Category'] = 'Medium'
Data.loc[(Data['AGE']>50), 'Age_Category'] = 'Old'
```

```python

Data
```

```python
def replace_zero(df, columns:List[str]):
    for column in columns:
        Data[column] = Data[column].replace(0, np.NaN)
    return df
Data = replace_zero(Data,['BMI','GLUCOSE','BLOODPRESSURE','SKINTHICKNESS','INSULIN'])
Data['BLOODPRESSURE'].fillna(Data['BLOODPRESSURE'].mean(),inplace = True)
Data['SKINTHICKNESS'].fillna(Data['SKINTHICKNESS'].mean(),inplace = True)
Data['INSULIN'].fillna(Data['INSULIN'].mean(),inplace = True)
Data['GLUCOSE'].fillna(Data['GLUCOSE'].mean(),inplace = True)
Data
```
